# Find All Anagrams in a String

Given two strings s and p, return an array of all the start indices of p's anagrams in s. You may return the answer in any order.

 

Example 1:

Input: s = "cbaebabacd", p = "abc"
Output: [0,6]
Explanation:
The substring with start index = 0 is "cba", which is an anagram of "abc".
The substring with start index = 6 is "bac", which is an anagram of "abc".
Example 2:

Input: s = "abab", p = "ab"
Output: [0,1,2]
Explanation:
The substring with start index = 0 is "ab", which is an anagram of "ab".
The substring with start index = 1 is "ba", which is an anagram of "ab".
The substring with start index = 2 is "ab", which is an anagram of "ab".
 

Constraints:

1 <= s.length, p.length <= 3 * 104
s and p consist of lowercase English letters

```python
class Solution:
    def findAnagrams(self,string, pattern):

        k = len(pattern)           #Window Size == pattern length
        result = []

        dicts = {}           
        for char in pattern:       #Convert pattern to dicts
            if char in dicts:
                dicts[char] += 1
            else:
                dicts[char] = 1

        count = len(dicts)         #Number of distinct chars in pattern

        leftPtr = 0
        for rgtPtr,char in enumerate(string):
            if char in dicts:                      #If char is in dicts
                dicts[char] -=1                    
                if dicts[char]==0:                 #If its value becomes zero
                    count -=1                          

            if rgtPtr-leftPtr+1 == k:               #Once window size is k
                if count==0:                        #If count is zero
                    result.append(leftPtr)          #Add leftPtr to result

                if string[leftPtr] in dicts:          #If leftPtr's char is in dict
                    if dicts[string[leftPtr]] == 0:   #If that char's val is zero
                        count +=1                     #Increment count
                    dicts[string[leftPtr]] +=1        #Increment the char's count value
                leftPtr+=1                            #Move to next window

        return result


'''
1) Create a dictionary whose keys are the characters in pattern and values are the num of occurences of those chars
Ex: pattern = aba 
    dicts = {a:2, b:1}
    count = 2           [Num of distinct chars in dict]

2) Start traversing the string char by char:
	 a) if the char is found in dict(pattern), decrement its value.
	 b) Check if its value becomes zero, decrement the count value.

3) When window size becomes k
	 a) If count is zero, meaning we found all the chars in pattern. Ex: dicts = {a:0, b=0}
	 b) Add the LeftPtr index to result

4) Now we have to move the window [but before that remove the char at leftPtr]

	a) Check if the char at leftPtr is in the dict
	b) Also Check if its value is zero, 
		if yes Increment the count Value 
		(Coz we decrement count only when chars value become zero(point 2b), so if it becomes greater than zero we shld increment it back)
		
	c) Also increment the char's value
	d) Finally move the window (leftPtr++) 
'''
```

```python
''''Firstly we will make window of the pattern given 'p' as ps, which would be consits of 26 number list initialize by 0.

Now make the sliding window of first 'n' length string same of pattern length 'pl', then start matching both the window after each operation if both window gets match then append the starting position of 'substring' window.
In each operation you go one step forward i.e. add coming element and subtract one from begining of the window, So that length of both window remains equal and you can compare at each operation.

'''

	class Solution:
	def findAnagrams(self, s: str, p: str) -> List[int]:

		if len(p)>len(s):
			return []

		ps = [0 for i in range(26)]
		pl=len(p)
		for letter in p:
			ps[ord(letter)-ord('a')]+=1

		substring = [0 for j in range(26)]
		for i in range(pl):
			substring[ord(s[i])-ord('a')]+=1

		res=[]
		if substring==ps:
			res.append(0)

		j=pl
		while j<len(s):
			substring[ord(s[j])-ord('a')]+=1
			substring[ord(s[j-pl])-ord('a')]-=1
			if substring==ps:
				res.append(j-pl+1)
			j+=1

		return res
```

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        if len(s) < len(p):
            return []
        data = {}
        for letter in p:
            if letter in data:
                data[letter] += 1
            else:
                data[letter] = 1

        left = 0
        right = len(p) - 1
        result = []
        search_data = defaultdict(int)

        for index in range(len(p) - 1):
            search_data[s[index]] += 1

        while right < len(s):
            search_data[s[right]] += 1

            if search_data == data:
                result.append(left)

            if search_data.get(s[left]) > 1:
                search_data[s[left]] -= 1
            else:
                del search_data[s[left]]
            left += 1
            right += 1

        return result
```

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        sh = [hash(c) for c in s]
        psum = sum([hash(c) for c in p])        
        # becomes: find all sliding windows of length len(p)
        # whose sum = psum        
        start_indices = []
        for i in range(0, len(s) - len(p) + 1):
            if sum(sh[i:i+len(p)]) == psum: start_indices.append(i)
        return start_indices
```
