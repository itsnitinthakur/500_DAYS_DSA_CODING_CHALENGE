# Find Peak Element

A peak element is an element that is strictly greater than its neighbors.

Given an integer array nums, find a peak element, and return its index. If the array contains multiple peaks, return the index to any of the peaks.

You may imagine that nums[-1] = nums[n] = -âˆž.

 

Example 1:

Input: nums = [1,2,3,1]
Output: 2
Explanation: 3 is a peak element and your function should return the index number 2.
Example 2:

Input: nums = [1,2,1,3,5,6,4]
Output: 5
Explanation: Your function can return either index number 1 where the peak element is 2, or index number 5 where the peak element is 6.
 

Constraints:

1 <= nums.length <= 1000
-231 <= nums[i] <= 231 - 1
nums[i] != nums[i + 1] for all valid i.
 

Follow up: Could you implement a solution with logarithmic complexity?

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        m = max(nums)
        if len(nums)<3:
            return nums.index(m)
        for i in range(1,len(nums)):
            if nums[i-1]<nums[i] and nums[i]> nums[i+1]:
                return i
        return nums.index(m)
        
```

It is guaranteed that the answer will fit in a 32-bit integer.

A subarray is a contiguous subsequence of the array.

### Explanation
The idea is to loop through the input and multiply each given element with the current product if its value is not 0. We also keep track of the highest negative product such that we can divide the current product, in case it is negative, by it to obtain the highest positive product possible among all subarrays with the given element as their last one. If the given element is 0, we will reset the current and the maximum negative product.

In summary, we always build the product of all non 0 elements and, if it becomes negative, we divide it by the maximum negative product. The current product is the product of all elements within a subarray which starts after the last encountered 0 or at the beginning of the input array and ranges up to the currently given element. The maximum negative product is the product of a prefix subarray which has the same starting element but covers less elements. It has the highest negative product among all prefix subarrays.

```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        min_product = None
        current_product = 1
        output = nums[0]
        
        for number in nums:
            output = max(output, number)
            
            if number == 0:
                current_product = 1
                min_product = None
            else:
                current_product *= number
                output = max(output, current_product)
                
                if min_product is not None:
                    other_product = current_product // min_product
                    output = max(output, other_product)
                
                if current_product < 0 and (min_product is None or current_product > min_product):
                    min_product = current_product

        return output
```

```python
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        dpmax = [None] * len(nums)
        dpmin = [None] * len(nums)
        
        ans = dpmax[0] = dpmin[0] = nums[0]
        
        for i in range(1, len(nums)):
            n = nums[i]
            
            if n >= 0:
                dpmax[i] = max(n, dpmax[i-1] * n)
                dpmin[i] = min(n, dpmin[i-1] * n)
            else:
                dpmin[i] = min(dpmax[i-1] * n, n)
                dpmax[i] = max(dpmin[i-1] * n, n)
                
            ans = max(ans, dpmax[i])
                
        return ans
```

# Minimum Size Subarray Sum

Given an array of positive integers nums and a positive integer target, return the minimal length of a contiguous subarray [numsl, numsl+1, ..., numsr-1, numsr] of which the sum is greater than or equal to target. If there is no such subarray, return 0 instead.

Output: 2

Example 3:

Input: target = 11, nums = [1,1,1,1,1,1,1,1]
Output: 0
 

Constraints:

1 <= nums[i] <= 105

```python
class Solution:
    def minSubArrayLen(self, tar: int, nums: List[int]) -> int:
        # pre = [0 for i in range(len(nums)+1)]
        # for i in range(1, len(nums)+1):
        #     pre[i] = pre[i-1]+nums[i-1]
        
        # if pre[-1] < tar: return 0
        
        low = 0
        high = 0
        sumi = 0
        ans = len(nums)+1
        while 1:
            if sumi < tar:
                if high >= len(nums):
                    break
                sumi += nums[high]
                high += 1
            else:
                if low >= len(nums):
                    break
                ans = min(ans, high-low)
                sumi -= nums[low]
                low += 1
        if ans > len(nums):
            return 0
        return ans
```
