# Sliding Window Maximum

You are given an array of integers nums, there is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the k numbers in the window. Each time the sliding window moves right by one position.

Return the max sliding window.

 

Example 1:

Input: nums = [1,3,-1,-3,5,3,6,7], k = 3
Output: [3,3,5,5,6,7]
Explanation: 
Window position                Max
---------------               -----
[1  3  -1] -3  5  3  6  7       3
 1 [3  -1  -3] 5  3  6  7       3
 1  3 [-1  -3  5] 3  6  7       5
 1  3  -1 [-3  5  3] 6  7       5
 1  3  -1  -3 [5  3  6] 7       6
 1  3  -1  -3  5 [3  6  7]      7
Example 2:

Input: nums = [1], k = 1
Output: [1]
Example 3:

Input: nums = [1,-1], k = 1
Output: [1,-1]
Example 4:

Input: nums = [9,11], k = 2
Output: [11]
Example 5:

Input: nums = [4,-2], k = 2
Output: [4]
 

Constraints:

1 <= nums.length <= 105
-104 <= nums[i] <= 104
1 <= k <= nums.length

```python
'''
Here the idea is maintaining this 'window' only contains 'meaningful' numbers.
If a number is smaller than another number that occurs later than it in this window,
then we consider it's not meaningful, because it will never have chance to be the max
'''
from collections import deque

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        def pushToWindow(window: deque, indice: int) -> None:
            if window and window[0] == indice - k:
                window.popleft()
            while window and nums[indice] > nums[window[-1]]:
                window.pop()
            window.append(indice)

        window = deque()
        res = []
        for i in range(len(nums)):
            pushToWindow(window, i)
            if i >= k - 1:
                max = nums[window[0]]
                res.append(max)
        return res
```

```python
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        q  = MonoQue()
        res = []
        for i, n in enumerate(nums):
            if i<k: 
                # initialize the queue:
                q.append(n)
            else:
                res.append(q.max())
                q.append(n) # move window right edge 
                q.popleft(nums[i-k]) # move window left edge
        res.append(q.max())
        return res
    
class MonoQue:
    """A monotonic queue object. It has the following property:
        (1) Items inside the queue preserve the order of appending.
        (2) Items inside the queue is non-increasing.
    """
    
    def __init__(self):
        from collections import deque
        self.q = deque()
        
    def append(self, n):
        while self.q and self.q[-1]<n:
            self.q.pop() # pop all elements that are smaller than n
        self.q.append(n)
    
    def popleft(self, n):
        # if the first element of the queue equals to n, pop it.
        if self.q[0]==n:
            self.q.popleft()
    
    def max(self):
        # the max of the queue is the first element.
        return self.q[0]
```

```python
class Heap:
    def __init__(self, arr):
        self.arr = sorted(arr)
        
    def insert(self, x):        
        arr = self.arr
        n = len(arr)
        if n == 0:
            self.arr.append(x)
            return
        
        if x >= arr[-1]:
            arr.append(x)
            return
        
        if x <= arr[0]:
            arr.insert(0, x)
            return 
        
        left, right = 0 , n - 1
        while left <= right:
            mid = left + (right - left) // 2
            if arr[mid] == x:
                arr.insert(mid, x)
                return
            elif arr[mid] < x and arr[mid + 1] > x:
                arr.insert(mid + 1, x)
                return
            elif arr[mid] > x and arr[mid - 1] < x:
                arr.insert(mid, x)
                return
            elif x < arr[mid]:
                right = mid - 1
            else:
                left = mid + 1
        
    def remove(self, x):
        arr = self.arr
        n = len(arr)
        if n == 0:
            return
        
        left, right = 0, n - 1
        while left <= right:
            mid = left + (right - left) // 2
            if arr[mid] == x:
                arr.pop(mid)
                return
            elif x < arr[mid]:
                right = mid - 1
            else:
                left = mid + 1
                
    def top(self):
        return self.arr[-1]
    
class Solution:                        
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        n = len(nums)
        heap = Heap(nums[:k])  # O(k log k) sort, one time        
        maxvals = [heap.top()]        
        for i in range(n - k):            
            heap.remove(nums[i])   # O(log n) remove
            heap.insert(nums[i + k])   # O(log n) insert 
            maxvals.append(heap.top())   # O(1) max
        return maxvals
```

```python
# since we want a data structure in which we can append at the end, remove from the front and also remove from the end, we would want a double linked list. python gives deque which is implemented in exactly similar way. 
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        queue = deque()
        result = []
        start = 0
        for end, num in enumerate(nums):
            # remove elements from the right as long as the number is greater than the elements 
            # because as soon as find a number greater than the previous number, the previous number
            # can never be the max. so we remove all lesser. 
            while queue and nums[queue[-1]] <= num:
                queue.pop()
            # we append new numbers as soon as we find them. 
            queue.append(end)
            # if the window is of size k + 1 now we remove from the front element of the queue if the front element is same as start. and then we move start to right by 1 to maintain the window of size k. 
            # ex: if end = 4, and start = 0 and k = 4, end - start = 4 means we are out of the window. 
            
            
            if queue and end - start == k:
                if queue[0] == start:
                    queue.popleft()
                start += 1
                
            # here now we are in the window of size k. so now we can append to our result the front most element in result
            if queue and end - start == k - 1:
                result.append(nums[queue[0]])
        return result
```

```python
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        dq,result=deque(),[]
        #the invariant is dq[0] is always the max in the window of k elements, dq might contain at most k elements and at least 1 elements at each index i(after append(i))
        def maintain_dq_invariant(i):
            if dq and dq[0]==i-k:
                dq.popleft()
            while dq and nums[dq[-1]]<=nums[i]: #if equal, we still pop it as we need only one max in the window
                dq.pop()
        for i in range(len(nums)):
            maintain_dq_invariant(i)
            dq.append(i)
            if (i>=k-1) :
                result.append(nums[dq[0]])
        return result 
```

```python
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: 'List[int]', k: 'int') -> 'List[int]':
        window = deque()
        out = []
        
        #deal with first k values
        for i in range(k):
            val = nums[i]
            while window and window[-1][0] <= val:
                window.pop()
            window.append((val,i))
        out.append(window[0][0])   
        
        #deal with values from k to end of list
        for i in range(k, len(nums)):
            #remove values that are too old
            while window and window[0][1] <= i-k:
                window.popleft()
            val = nums[i]
            
            #remove values that are smaller than new value
            while window and window[-1][0] <= val:
                window.pop()
                
            #add new value
            window.append((val,i))
            out.append(window[0][0])
            
        return out
```
