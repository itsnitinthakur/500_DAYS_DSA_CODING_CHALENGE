# Shortest Unsorted Continuous Subarray

Given an integer array nums, you need to find one continuous subarray that if you only sort this subarray in ascending order, then the whole array will be sorted in ascending order.

Return the shortest such subarray and output its length.

 

Example 1:

Input: nums = [2,6,4,8,10,9,15]
Output: 5
Explanation: You need to sort [6, 4, 8, 10, 9] in ascending order to make the whole array sorted in ascending order.
Example 2:

Input: nums = [1,2,3,4]
Output: 0
Example 3:

Input: nums = [1]
Output: 0
 

Constraints:

1 <= nums.length <= 104
-105 <= nums[i] <= 105

```python
class Solution:
    def findUnsortedSubarray(self, nums: List[int]) -> int:
        sortedArray=sorted(nums) 
        if nums==sortedArray:
            return 0
        i,c=0,0
        while nums[i]==sortedArray[i] and i<len(nums):
            i+=1
            c+=1
        j=len(nums)-1
        while nums[j]==sortedArray[j] and j>-1:
            j-=1
            c+=1
        return len(nums)-c
```

```python
class Solution:
    def findUnsortedSubarray(self, arr: List[int]) -> int:
        max_sofar = -float('inf')
        min_sofar = float('inf')
        n = m = 0
        twisted = False
        for i in range(len(arr)):
            # find n
            # keep track of max
            max_sofar = max(max_sofar, arr[i])
            # the last element that less than max is 'n'
            if arr[i] < max_sofar : n = i

            # find m from min
            # get min since twisted (number go down)
            if i: # skip first element
                if arr[i] < arr[i-1]: # twist
                    twisted = True
            # min track
            if twisted:
                min_sofar = min(arr[i],min_sofar)

        # find m where m this the first position more than min
        for i in range(len(arr)):
            if arr[i] > min_sofar:
                m = i
                break
        if n :
            return n-m+1
        else:
            return 0
```

```python
class Solution:
    def findUnsortedSubarray(self, nums: List[int]) -> int:
        n = len(nums)
        # find first break from start
        b1 = 0
        while b1 < n - 1 and nums[b1] <= nums[b1 + 1]:
            b1 += 1
            
        # already sorted, not subarray sorting needed
        if b1 == n - 1:
            return 0
        
        # find first break from end
        b2 = n - 1
        while b2 > 0 and nums[b2 - 1] <= nums[b2]:
            b2 -= 1
        
        # find min/max in [b1-b2] range
        if b1 == b2 - 1:
            min_range = nums[b2]
            max_range = nums[b1]
        else:
            min_range, max_range = float('inf'), float('-inf')
            for i in range(b1, b2 + 1):
                min_range = min(nums[i], min_range)
                max_range = max(nums[i], max_range)
        
        # move b1 back
        while b1 >= 0 and nums[b1] > min_range:
            b1 -= 1
        
        # move b2 forward
        while b2 < n and nums[b2] < max_range:
            b2 += 1
        return b2 - b1 - 1
```
